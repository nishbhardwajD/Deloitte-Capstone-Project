```python
df1 = pd.read_csv('HR_Hiring_details_transaction.csv')
df2 = pd.read_csv('Interview_Dates.csv')
df3 = pd.read_csv('Joining_Status.csv')
```

```python
df4 = pd.merge(df1, df2, on = "Candidate_Ref", how = "inner")
```

```python
df5 = pd.merge(df4, df3, on = "Candidate_Ref", how = "inner")
```

```python
df5
```

```python
df6 = pd.read_csv('Domicile_Master.csv')

df5.rename(columns = {'Domicile_Id':'Domicile_ID'}, inplace = True)

df7 = pd.merge(df5, df6, on = "Domicile_ID", how = "inner")

df8 = pd.read_csv('LOB_Master.csv')

df9 = pd.merge(df7, df8, on = "LOB_Id", how = "inner")

df10 = pd.read_csv('Location_Master.csv')

df11 = pd.merge(df9, df10, on = "Location_ID", how = "inner")

df12 = pd.read_csv('Postal_Code_Master.csv')

df13 = pd.merge(df11, df12, on = "Location_ID", how = "inner")

df14 = pd.read_csv('Region_Location_Mapping.csv')

df15 = pd.merge(df13, df14, on = "Location_ID", how = "inner")

df16 = pd.read_csv('Region_Master.csv')

df17 = pd.merge(df15, df16, on = "Region_ID", how = "inner")
```
